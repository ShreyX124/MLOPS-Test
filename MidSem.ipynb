```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Load the datasets
train_data = pd.read_csv('train.csv', encoding='latin1')
test_data = pd.read_csv('test.csv', encoding='latin1')

# 1. Basic Information
print("Training Data Info:")
print(train_data.info())
print("\nTest Data Info:")
print(test_data.info())

print("\nFirst 5 rows of Training Data:")
print(train_data.head())

# 2. Check for Missing Data
print("\nMissing Values in Training Data:")
print(train_data.isnull().sum())

print("\nMissing Values in Test Data:")
print(test_data.isnull().sum())

# Visualize Missing Data
plt.figure(figsize=(10, 6))
sns.heatmap(train_data.isnull(), cbar=False, cmap='viridis')
plt.title("Missing Data Heatmap - Training Data")
plt.show()

# 3. Statistical Summary
print("\nStatistical Summary of Numerical Features (Training Data):")
print(train_data.describe())

# 4. Sentiment Distribution
plt.figure(figsize=(8, 5))
sns.countplot(data=train_data, x='sentiment', order=train_data['sentiment'].value_counts().index, palette="viridis")
plt.title("Sentiment Distribution")
plt.xlabel("Sentiment")
plt.ylabel("Count")
plt.show()

# 5. Text Length Analysis
train_data['text_length'] = train_data['text'].fillna("").apply(len)
plt.figure(figsize=(10, 6))
sns.histplot(train_data['text_length'], bins=30, kde=True, color='blue')
plt.title("Distribution of Text Lengths")
plt.xlabel("Text Length")
plt.ylabel("Frequency")
plt.show()

# 6. Correlations in Numerical Data
numerical_columns = ['Population -2020', 'Land Area (Km²)', 'Density (P/Km²)']
correlation_matrix = train_data[numerical_columns].corr()
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Matrix - Numerical Features")
plt.show()

# 7. Sentiment vs. Numerical Features
plt.figure(figsize=(10, 6))
sns.boxplot(data=train_data, x='sentiment', y='Population -2020', palette="viridis")
plt.title("Population by Sentiment")
plt.show()

plt.figure(figsize=(10, 6))
sns.boxplot(data=train_data, x='sentiment', y='Density (P/Km²)', palette="magma")
plt.title("Density by Sentiment")
plt.show()

# 8. Time of Tweet Analysis
plt.figure(figsize=(8, 5))
sns.countplot(data=train_data, x='Time of Tweet', order=train_data['Time of Tweet'].value_counts().index, palette="cool")
plt.title("Frequency of Tweets by Time")
plt.xlabel("Time of Tweet")
plt.ylabel("Count")
plt.show()

# 9. Age of User Analysis
plt.figure(figsize=(8, 5))
sns.countplot(data=train_data, x='Age of User', order=train_data['Age of User'].value_counts().index, palette="cividis")
plt.title("Distribution of User Ages")
plt.xlabel("Age of User")
plt.ylabel("Count")
plt.show()

# 10. Country Analysis
top_countries = train_data['Country'].value_counts().head(10)
plt.figure(figsize=(10, 6))
sns.barplot(x=top_countries.values, y=top_countries.index, palette="viridis")
plt.title("Top 10 Countries by Tweet Count")
plt.xlabel("Count")
plt.ylabel("Country")
plt.show()
```

```python
# Importing necessary libraries
import pandas as pd
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense, Dropout
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt

# Load datasets
train_data = pd.read_csv('train.csv', encoding='latin1')
test_data = pd.read_csv('test.csv', encoding='latin1')

# Preprocessing: Extracting relevant columns
X = train_data['text'].fillna('')  # Fill missing values with empty string
y = train_data['sentiment']

# Convert non-string values to strings (in case there are any float or numeric values)
X = X.astype(str)

# Encoding the target labels (neutral=0, negative=1, positive=2)
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)

# Tokenizing the text data
max_words = 5000  # Vocabulary size
max_len = 100     # Max sentence length
tokenizer = Tokenizer(num_words=max_words, lower=True)
tokenizer.fit_on_texts(X)
X_seq = tokenizer.texts_to_sequences(X)
X_pad = pad_sequences(X_seq, maxlen=max_len)

# Split data into training and validation sets
X_train, X_val, y_train, y_val = train_test_split(
    X_pad, y_encoded, test_size=0.2, random_state=42, stratify=y_encoded
)

# Define the LSTM model
model = Sequential([
    Embedding(input_dim=max_words, output_dim=128, input_length=max_len),
    LSTM(64, dropout=0.2, recurrent_dropout=0.2),
    Dense(3, activation='softmax')  # Output layer for 3 classes (neutral, negative, positive)
])

# Compile the model
model.compile(
    loss='sparse_categorical_crossentropy',
    optimizer='adam',
    metrics=['accuracy']
)

# Train the model
history = model.fit(
    X_train, y_train,
    epochs=5,
    batch_size=64,
    validation_data=(X_val, y_val),
    verbose=2
)

# Evaluate the model on validation data
val_loss, val_accuracy = model.evaluate(X_val, y_val, verbose=0)
print(f"Validation Accuracy: {val_accuracy * 100:.2f}%")

# Plot training and validation accuracy
plt.plot(history.history['accuracy'], label='Train Accuracy')
plt.plot(history.history['val_accuracy'], label='Val Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.title('Training and Validation Accuracy')
plt.legend()
plt.show()

# Preprocess the test data
X_test_seq = tokenizer.texts_to_sequences(test_data['text'].fillna('').astype(str))  # Handle missing values and ensure strings
X_test_pad = pad_sequences(X_test_seq, maxlen=max_len)

# Make predictions on the test set
test_predictions = model.predict(X_test_pad)
test_predictions_labels = tf.argmax(test_predictions, axis=1)

# Decode predictions to original sentiment labels
test_data['predicted_sentiment'] = label_encoder.inverse_transform(test_predictions_labels.numpy())

# Save the predictions to a CSV file
output_path = 'test_predictions_sentiment.csv'
test_data[['text', 'predicted_sentiment']].to_csv(output_path, index=False)
print(f"Test Predictions saved to {output_path}")
```
